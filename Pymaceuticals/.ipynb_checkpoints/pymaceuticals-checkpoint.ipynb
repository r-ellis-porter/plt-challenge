### Analysis

- There is a strong correlation (r > 0.7) between mouse weight and the average tumor volume for mice on the Capomulin regimen.

- The strong correlation between mouse weight and the average tumor volume for mice on the Capomulin regimen is positive. 

- The distribution of final tumor volumes for mice on the Capomulin and Ramicane regimens was lower than the distribution of final tumor volumes for mice on the Infubinol and Ceftamin regimens.


```python
# Dependencies and Setup

import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files

mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results

mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame

df = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview

df
```

```python
# Checking the number of mice.

df["Mouse ID"].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

df[df.duplicated(subset=["Mouse ID", "Timepoint"])]["Mouse ID"].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

df.loc[df["Mouse ID"] == "g989"]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

df = df.drop(df[df["Mouse ID"] == "g989"].index)
```

```python
# Checking the number of mice in the clean DataFrame.

df["Mouse ID"].nunique()
```

## Summary Statistics

```python
## Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

df_group = df.groupby(["Drug Regimen"])

mean = df_group["Tumor Volume (mm3)"].mean()
median = df_group["Tumor Volume (mm3)"].median()
var = df_group["Tumor Volume (mm3)"].var()
sd = df_group["Tumor Volume (mm3)"].std()
error = df_group["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary DataFrame.

per_drug_summary = pd.DataFrame({
    "Mean Tumor Volume": mean,
    "Median Tumor Volume": median,
    "Tumor Volume Variance": var,
    "Tumor Volume Std. Dev.": sd,
    "Tumor Volume Std. Err.": error
})

per_drug_summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line

agg_summary = df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)": ["mean", "median", "var", "std", "sem"]})

agg_summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
mouse_group = df.groupby(["Mouse ID"])["Timepoint"].max()
mouse_data = mouse_group.reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
mouse_data_merge = pd.merge(mouse_data, df, how="left", on=["Mouse ID", "Timepoint"])
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    # add subset 

for drug in drugs:
    
    vol = mouse_data_merge[mouse_data_merge["Drug Regimen"] == drug]["Tumor Volume (mm3)"]
    
    quartiles = vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    
    tumor_vol.append(vol)
    
    # Determine outliers using upper and lower bounds
    
    lowerb = lowerq - (1.5*iqr)
    upperb = upperq + (1.5*iqr)
    
    outlier = vol.loc[(vol < lowerb) | (vol > upperb)]
    
    print(f"{drug}'s potential outliers: {outlier}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

outliers = dict(markerfacecolor='red',markersize=12)

plt.boxplot(tumor_vol, labels=drugs, flierprops=outliers)
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

## Line and Scatter Plots

```python
solo_data = df.loc[df["Mouse ID"] == "l509"]
solo_2 = solo_data.loc[solo_data["Drug Regimen"] == "Capomulin"]
```

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
solo_tp = solo_2["Timepoint"]
solo_vol = solo_2["Tumor Volume (mm3)"]

plt.plot(solo_tp, solo_vol)
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin Treatment of Mouse l509")

plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

cap_data = df.loc[df["Drug Regimen"] == "Capomulin"]
cap_group = cap_data.groupby(["Mouse ID"])[["Weight (g)", "Tumor Volume (mm3)"]].mean()
cap_weight = cap_group["Weight (g)"]
cap_vol = cap_group["Tumor Volume (mm3)"]

plt.scatter(cap_weight, cap_vol)

plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

plt.scatter(cap_weight, cap_vol)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

correlation = st.pearsonr(cap_weight,cap_vol)

(m, b, r, p, se) = linregress(cap_weight, cap_vol)
r_values = m * cap_weight + b
plt.plot(cap_weight,r_values,"red")

plt.annotate(f"RÂ²: {round(r**2, 6)}",(20, 36), fontsize=15,color="red")

print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}")
plt.show()
```
